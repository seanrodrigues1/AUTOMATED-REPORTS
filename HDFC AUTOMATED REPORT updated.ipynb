```python
def func(row):
    
    if ((row['System Score']>=50)  and (row['user_rating']==4)):
        val="Relevant,Relevant"
        
    
    elif ((row['System Score']>=50)  and (row['user_rating']==2)) :
        val = "Relevant,Not-Relevant"
    
    elif ((row['System Score']<50)  and (row['user_rating']==4)):
        val="Not-Relevant,Relevant"
    
    elif ((row['System Score']<50)  and (row['user_rating']==2)) :
    
        val ="Not-Relevant,Not-Relevant"
    else:
        val=""
            
    return val

df_final['System-User relevance'] = df_final.apply(func, axis=1)
```
